```python
import time
import queue
import threading

fila = queue.Queue()

def produtor():
    for i in range(3):
        evento = f"evento-{i}"
        print("[Produtor] Enviando:", evento)
        fila.put(evento)
        time.sleep(0.3)

def worker():
    while True:
        evento = fila.get()
        print("[Worker] Processando:", evento)
        time.sleep(0.5)
        fila.task_done()

threading.Thread(target=worker, daemon=True).start()
produtor()
fila.join()
```

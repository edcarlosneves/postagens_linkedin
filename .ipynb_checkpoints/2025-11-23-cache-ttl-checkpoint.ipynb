```python
import time
from functools import wraps

def cache_ttl(ttl):
    def decorator(func):
        cache = {}
        @wraps(func)
        def wrapper(*args):
            agora = time.time()
            if args in cache:
                valor, timestamp = cache[args]
                if agora - timestamp < ttl:
                    return valor
            resultado = func(*args)
            cache[args] = (resultado, agora)
            return resultado
        return wrapper
    return decorator

@cache_ttl(3)
def somar(a, b):
    print("Executando cálculo...")
    time.sleep(1)
    return a + b

start_time = time.time()
print(f"Sem Cache | Resultado da Operação={somar(1, 2)} | Tempo de execução={time.time() - start_time}")

start_time = time.time()
print(f"Com Cache | Resultado da Operação={somar(1, 2)} | Tempo de execução={time.time() - start_time}")
```

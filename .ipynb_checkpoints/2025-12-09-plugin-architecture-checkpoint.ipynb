# main.py

```python
import importlib


def carregar_plugin(nome):
    modulo = importlib.import_module(nome)
    return modulo.Plugin()


plugin = carregar_plugin("plugin_soma")
print(plugin.executar(2, 3))
```

# plugin_soma.py

```python
class Plugin:
    def executar(self, a, b):
        return a + b
```

```python
from pydantic import BaseModel, Field, ValidationError

class SensorData(BaseModel):
    temperatura: float = Field(gt=-100, lt=500)
    pressao: float = Field(gt=0)
    unidade: str

try:
    dados = SensorData(temperatura=85.3, pressao=10.2, unidade="SI")
    print(dados)
    print("----")
    dados = SensorData(temperatura=85.3, pressao=10.2, unidade=10)
except ValidationError as e:
    print(e)
```

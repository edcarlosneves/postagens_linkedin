```python
import queue
import random

fila = queue.Queue()
dlq = queue.Queue()


def worker():
    while not fila.empty():
        msg = fila.get()
        try:
            if random.random() < 0.4:
                raise Exception("Falha ao processar")
            print("Processado:", msg)
        except:
            print("Enviado para DLQ:", msg)
            dlq.put(msg)
        fila.task_done()


for i in range(10):
    fila.put(f"msg-{i}")

worker()

print("Mensagens na DLQ:")
while not dlq.empty():
    print(dlq.get())
```

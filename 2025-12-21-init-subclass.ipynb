```python
class EntidadeBanco:
    def __init_subclass__(cls, **kwargs):
        super().__init_subclass__(**kwargs)

        if not hasattr(cls, "tabela"):
            raise TypeError(
                f"A classe '{cls.__name__}' precisa definir o atributo 'tabela'."
            )

        print(f"Classe '{cls.__name__}' validada. Tabela: '{cls.tabela}'")
```

```python
class Cliente(EntidadeBanco):
    tabela = "clientes"
```

```python
class Pedido(EntidadeBanco):
    pass
```

```python
import time
import threading
from random import randint


class Scheduler:
    def __init__(self):
        self.jobs = []

    def add_job(self, func, intervalo):
        self.jobs.append((func, intervalo))

    def start(self):
        for func, intervalo in self.jobs:
            thread = threading.Thread(target=self._executar, args=(func, intervalo))
            thread.daemon = True
            thread.start()

    def _executar(self, func, intervalo):
        while True:
            func()
            time.sleep(intervalo)


def ler_sensor():
    temperatura = randint(0, 100)
    print(f"Lendo sensor em {time.ctime(time.time())} | Temperatura -> {temperatura}")


def gerar_relatorio():
    print(f"Gerando relatório em {time.ctime(time.time())}")


scheduler = Scheduler()
scheduler.add_job(ler_sensor, intervalo=2)
scheduler.add_job(gerar_relatorio, intervalo=2)

scheduler.start()

while True:
    time.sleep(1)
```

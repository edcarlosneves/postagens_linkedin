```python
import threading
import time
import random


class Bulkhead:
    def __init__(self, limite_concorrencia):
        self.semaforo = threading.Semaphore(limite_concorrencia)

    def executar(self, func):
        with self.semaforo:
            return func()


bulk = Bulkhead(limite_concorrencia=3)


def tarefa(id):
    print(f"Tarefa {id} iniciou")
    time.sleep(random.uniform(0.5, 1.5))
    print(f"Tarefa {id} finalizou")


threads = []
for i in range(5):
    t = threading.Thread(target=bulk.executar, args=(lambda i=i: tarefa(i),))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```
